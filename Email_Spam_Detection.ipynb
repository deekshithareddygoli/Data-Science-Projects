1. Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

2. Data Collection and Pre-Processing

```python
#Loading the data from csv file to a data frame
spam_data = pd.read_csv('/content/spam.csv', encoding='latin-1')
spam_data
```

```python
spam_data.head()
```

```python
spam_data.shape
```

```python
spam_data.info()
```

```python
#Checking for null or missing values
spam_data.isnull().sum()
```

```python
# Let's drop the Unnamed columns
spam_data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
spam_data.isnull().sum()
```

```python
spam_data.head()
```

```python
spam_data.columns
```

```python
spam_data.columns = ['Category', 'Message']
spam_data.head()
```

```python
spam_data.describe()
```

Label Encoding

```python
spam_data.replace({'Category': {'spam': 0, 'ham': 1}}, inplace=True)
spam_data.head()
```

Splitting the data and target

```python
X = spam_data['Message']
Y = spam_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python
# Transform the text data to feature vectors that can be used as input to Logistic Regression
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Model Training

Logistic Regression

```python
log_lin_model = LogisticRegression()
```

```python
log_lin_model.fit(X_train_features, Y_train)
```

Model Evaluation

```python
# Prediction on training data
training_data_prediction = log_lin_model.predict(X_train_features)
accuracy_training_data_prediction = accuracy_score(Y_train, training_data_prediction)
```

```python
print('Accuracy on training data : ', accuracy_training_data_prediction)
```

```python
# Prediction on test data
test_data_prediction = log_lin_model.predict(X_test_features)
accuracy_test_data_prediction = accuracy_score(Y_test, test_data_prediction)
```

```python
print('Accuracy on test data : ', accuracy_test_data_prediction)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = log_lin_model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
